```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
import pandas as pd
import random
import scipy.optimize as opt
from sklearn.metrics import classification_report, accuracy_score#这个包是评价报告
```

```python
# =========== Part 1: Loading and Visualizing Data =============
loadmat = loadmat('ex3data1.mat')
X = loadmat['X']
y = loadmat['y']
rand_indices = random.sample(np.arange(5000).tolist(), 100)
sel = X[rand_indices,:]
X.shape, y.shape
```

```python
def displayData(X):
    #  实际就是把（100,400) ，每行转换成20*20 像素的小矩阵即一个数字，100行共100个数字
    example_width = int(round(np.sqrt(X.shape[1])))
    [m,n] = X.shape
    example_height = int(n / example_width)
    display_rows = int(np.floor(np.sqrt(m)))
    display_cols = int(np.ceil(m / display_rows))
    pad = 1
    display_array = pd.DataFrame(-np.ones([pad + display_rows * (example_height + pad), pad + display_cols * \
                                           (example_width + pad)]))
    curr_ex = 0
    for j in range(display_rows):
        for i in range(display_cols):
            max_val = max(abs(X[curr_ex,:]))
            display_array.iloc[pad + j*(example_height+pad) + np.arange(example_height),\
                         pad + i*(example_width+pad) + np.arange(example_width)] = \
            X[curr_ex,:].reshape(example_height, example_width).T / max_val  #  和 matlab的reshap不同，需要加转置
            curr_ex += 1

    plt.imshow(display_array)
displayData(sel)
```

```python
# ============ Part 2: Vectorize Logistic Regression ============
num_labels = 10 #  共10个输出单元
def sigmoid(z):
    return 1 / (1+pow(np.e, -z))

def model(X,theta):
    return sigmoid(np.dot(X, theta))

def regularized_cost(theta, X, y, lamda, m):
    theta = theta.reshape(len(theta),1)
    ta = np.concatenate((np.array([[0]]), theta[1:]), axis=0)  # (401, 1)
    cost = -(y*np.log(model(X, theta) + 1e-5) + (1-y)*np.log( 1 - model(X, theta))).mean()  + lamda/(2*m)  * pow(theta[1:], 2).sum() # np.dot(ta.T, ta).sum().sum()
    return cost 

def regularized_grad(theta, X, y, lamda, m):
    theta = theta.reshape(len(theta),1)   # 使用minimize的时候会把theta变成一维的
    ta = np.concatenate((np.array([[0]]), theta[1:]), axis=0)  # (401, 1)
    grad = np.dot(X.T, (model(X, theta) - y))/m + lamda / m * ta    # model(X, theta).shape : (5000, 1)
    return grad

def descent(X, y, theta, m, alpha=0.01, thread=0.02, lamda=0.1 ):
    cost_lst = []
    iterations = 0
    k = 0
    total_samples = len(X)
    while True:
        iterations += 1
        cost = regularized_cost(theta, X, y, lamda, m)
        grad = regularized_grad(theta, X, y, lamda, m)
        cost_lst.append(cost)
        theta = theta - alpha * grad
        if iterations == 5000: break
        if np.linalg.norm(grad) < thread: break
#         if cost < 0.057252: break
    print(cost)
    plt.plot(np.arange(1,1+iterations), cost_lst)
    return theta, cost_lst, iterations
```

## 1.使用手写代码预测

```python
X = np.insert(X,0,1,axis=1)
```

```python
m = X.shape[0]
theta_lst = []
for unit_pos in range(1, num_labels+1):
    theta = np.zeros((401,1))
    [theta, cost_lst, iterations] = descent(X, (y==unit_pos), theta, m, alpha=0.1, thread=0.0005)
    theta_lst.append(theta.T)
all_theta = np.concatenate((theta_lst))  # (10, 401)
```

```python
# ================ Part 3: Predict for One-Vs-All ================
def predict(X, all_theta):
    C = model(X, all_theta.T)
    value = np.max(C, axis=1).reshape(len(C),1)  
    pos = np.argmax(C, axis=1).reshape(len(C),1)+1 # 每个样本进行多类预测，根据预测值取出最有可能分类
    return value, pos
[value, pos] = predict(X, all_theta)
accuracy = (y == pos).mean() * 100
accuracy
```

## 2.使用optimize包优化

```python
m = X.shape[0]
theta = np.zeros((401,1))
lamda = 1
theta_lst1 = []
for unit_pos in range(1, num_labels+1):
    res = opt.minimize(fun=regularized_cost,
                        x0=theta,
                        args = (X, (y==unit_pos), lamda, m),
                         method='TNC',
                         jac=regularized_grad,
                         options={'disp': False})
    theta_lst1.append([res.x])
#     print(regularized_cost(res.x, X, (y==unit_pos), lamda, m))
all_theta1 = np.concatenate((theta_lst1)) 

[value1, pos1] = predict(X, all_theta1)
accuracy = (y == pos1).mean() * 100
accuracy  # 使用optmize 速度快，精度也提高
```
